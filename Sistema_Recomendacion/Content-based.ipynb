```python
#reseteo índice
userNews = userNews.reset_index(drop=True)
#cols inecesarias
userWebTable = userNews.drop('new_id', 1).drop('Título', 1).drop('Web', 1)
userWebTable
```

```python
inputNews['rating']
```

```python
#calculo de pesos de cada web
userProfile = userWebTable.transpose().dot(inputNews['rating'])

userProfile
```

```python
webTable = newsWeb_df.set_index(newsWeb_df['new_id'])

webTable = webTable.drop('new_id', 1).drop('Título', 1).drop('Web', 1)
webTable.head()
```

```python
webTable.shape
```

```python
#Se multiplica web por el peso y nos quedamos con la media
recommendationTable_df = ((webTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head()
```

```python
#Orden descendente
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)

recommendationTable_df.head()
```
